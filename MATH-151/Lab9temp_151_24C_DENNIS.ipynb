```python
# Enter your code here.
x, l = sp.symbols('x l')
Area = x * l + ((sp.pi * (l / 2)**2)) / 2
display(Area)
```

#### 2b

```python
# Enter your code here.

# Perimeter Equation is: 150 = 2x + l + (pi * l / 2)
length = (2 * (150 - 2 * x)) / (2 + sp.pi)
Area2 = Area.subs(l, length)
Aprime = Area2.diff(x)
max = sp.solve(Aprime, x)
print('The function obtains a critical value at x =', max)
Aprime2 = Aprime.diff(x)
concavity = Aprime2.subs(x, max[0])
print(f'Since the concavity is {concavity}, the function is concave down at this point, meaning the function has a max at this value')
```
