### Question 2

#### 2a

```python
# P, r, t = sp.symbols('P, r, t')
P = sp.Symbol('P')
r = sp.Symbol('r')
t = sp.Symbol('t')

R = P * ( r/12 * (1 + r/12)**(12*t)) / ((1+r/12)**(12*t) - 1)
display(R)

# 2a-i: Monthly payment for 30 years
M_30 = R.subs([(P, 400_000), (r, 0.06), (t, 30)])
print("$", M_30)
# 2a-ii: Monthly payment for 12 years
M_12 = R.subs([(P, 400_000), (r, 0.06), (t, 12)])
print("$", M_12)
```

#### 2b

```python
# Calculate total for 30 year
T_30 = M_30 * 30*12
print("$", T_30)
# Calculate total for 12 year
T_12 = M_12 * 12*12
print("$", T_12)
```

#### 3a

```python
# put your code here
F_x = 12
F_y = 7

F_mag = (F_x*F_x + F_y*F_y)**.5
print('<', F_x, ',', F_y, '>')
print('Magnitude of F =', F_mag)
```

#### 3b

```python
d_x = 27-14
d_y = 12-3
print('<', d_x, ',', d_y, '>')
```

#### 3c

```python
d_mag = (d_x*d_x + d_y*d_y)**.5
print('Magnitude of d =', d_mag)
```

#### 3d

```python
# W = F * d = F.x * d.x + F.y * d.y
W = F_x * d_x + F_y * d_y
print("The work done by the force on the object =", W)
```

#### 3e

```python
# W = F * d = |F||d|cosθ
# cosθ = (F * d) / |F||d| = W / |F||d|
cos_theta = W / (F_mag * d_mag)
print("cosθ =", cos_theta)
```

#### 3f

```python
# put your code here
# θ = acos(cos_theta)
print("θ =", sp.acos(cos_theta), "radians")
```
