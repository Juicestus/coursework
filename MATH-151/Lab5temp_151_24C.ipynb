```python
import sympy as sp
from sympy.plotting import (plot,plot_parametric,plot_implicit)
```

### Question 1

#### 1a

```python
t = sp.Symbol('t')
k = sp.Symbol('k', real=True)
y = sp.cos(k * t)

diffeq = sp.Eq(sp.diff(y, t, 6) + 64 * y, 0)

solutions = sp.solve(diffeq, k)

for solution in solutions:
    display(sp.Eq(k, solution))
```

#### 1b

```python
A, B = sp.symbols('A B')
Y_k = A * sp.sin(k * t) + B * sp.cos(k * t)

for solution in solutions:
    Y = Y_k.subs(k, solution)
    # display(Y)
    subst = diffeq.subs(y, Y)
    display(subst)
    is_it_eq = sp.Eq(subst, diffeq)
    print("New equation with k solution substitued is a solution to the original differential:", is_it_eq)
```
