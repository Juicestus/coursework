### Question 2

#### 2a

```python
x, y = sp.symbols('x y')
LHS = y**2
RHS = x**3 + 3*x**2
dy_dx = sp.idiff(LHS - RHS, y, x)
display(dy_dx)
```

#### 2b

```python
slope = dy_dx.subs({x: 1, y: -2})
b = -2 - slope
# print(f'y = {slope}x + {b}')
display(sp.Eq(y, slope * x + b))
```

#### 2c

```python
htan = sp.solve([3*x*(x/2 + 1), LHS - RHS], [x,y])
print('There are horizontal tangents at', htan)

vtan = sp.solve([y, LHS - RHS], [x, y])
print('There are vertical tangents at', vtan)
```

#### 2d

```python
p = plot_implicit(LHS-RHS, (x, -4, 4), y, show = False, line_color = 'red')
hplot = sp.plot((-2, (x, -4, 4)),(2, (x, -4, 4)),(0,(x, -4, 4)), show = False, line_color = 'blue')
p.extend(hplot)
t = sp.Symbol('t')
vplot = plot_parametric((-3, t, (t, -4, 4)), (0, t, (t, -4, 4)), show = False, line_color = 'blue')
p.extend(vplot)
p.show()
```
