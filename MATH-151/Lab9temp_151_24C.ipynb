### Question 1

#### 1a

```python
# Check variables (z = theta)
z = sp.Symbol("z", real=True)

# Compute cross-sectional area (0 <= z <= pi/2)
A = (20 * 10*sp.sin(z)) + (10*sp.cos(z) * 10*sp.sin(z))
Ap = sp.diff(A, z)

for sol in sp.solve(Ap, z):
  if sol > 0:
    max = sol

display("Exact maximum solution (rad):", max)
display("Approximate maximum solution (rad):", max.evalf())
```

#### 1b

```python
# Verify point is a maximum
display("Lower bound (cm^2), A(0) =", A.subs(z, 0),)
display("Critical point (cm^2), A(1.196) =", A.subs(z, max).evalf())
display("Upper bound (cm^2), A(pi/2) =", A.subs(z, sp.pi / 2))
```

### Question 2

#### 2a

```python
x, l = sp.symbols('x l')
Area = x * l + ((sp.pi * (l / 2)**2)) / 2
display(Area)
```

#### 2b

```python
# Perimeter Equation is: 150 = 2x + l + (pi * l / 2)
length = (2 * (150 - 2 * x)) / (2 + sp.pi)
Area2 = Area.subs(l, length)
Aprime = Area2.diff(x)
max = sp.solve(Aprime, x)
print('The function obtains a critical value at x =', max)
Aprime2 = Aprime.diff(x)
concavity = Aprime2.subs(x, max[0])
print(f'Since the concavity is {concavity}, the function is concave down at this point, meaning the function has a max at this value')
```

#### 2c

```python
Area3 = Area2.subs(x, max[0])
display(Area3)
```

```python
x, C = sp.symbols('x c')

fpp = 8 / ( (x + 2) ** 4 )

fp = sp.integrate(fpp, x)
C = 16 - fp.subs(x, 0)
print("f'(x) is equal to") 
display(fp + C)

print()

f = sp.integrate(fp, x)
C = 9 - fp.subs(x, 0)
print("f(x) is equal to") 
display(f + C)
```

#### 3b

```python
x, c, d, C, D = sp.symbols('x c d C D')

fpp = 8 / ( (x + 2) ** 4 )

fp = sp.integrate(fpp, x) + c
f = sp.integrate(fp, x) + d

f1 = f.subs(x, 0) # no c, only in terms of d
f2 = f.subs(x, 8)

D = sp.solve(f1, d)[0] # there is only 1 solution 
C = sp.solve(f2.subs(d, D), c)[0] 

print('f(x) is equal to')
display(f.subs(d, D).subs(c, C))
```
