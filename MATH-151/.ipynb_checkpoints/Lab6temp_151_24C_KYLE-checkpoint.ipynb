### Question 3

#### 3a

```python
# Enter your code here

yn = sp.symbols("yn")
r = sp.Symbol("r", real=True)

# Equation 1: 50000 = yn * sp.E**(r * 2)
#                 0 = yn * sp.E**(r * 2) - 50000
# Equation 2: 60000 = yn * sp.E**(r * 6)
#                 0 = yn * sp.E**(r * 6) - 60000

vars = sp.solve([yn * sp.E**(r * 2) - 50000, yn * sp.E**(r * 6) - 60000], [yn, r])
yn = vars[0][0]
r = vars[0][1]
display(f"y0  =  {yn}")
display(f"r   =  {r}  =  {r.evalf()}")
```
