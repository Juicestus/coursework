### Question 1

#### 1a

```python
# Define vars and functions
x = sp.Symbol("x", real=True)
f = (-x**7 - 2*x**6 + 16*x**5 + 20*x**4 - 64*x**3 + 36*x**2 - 72) / 100

# Find critical values
fp = sp.diff(f, x)
crits = []
for crit in sp.solve(fp, x):
  crits.append(crit.evalf())

print("Critical values at x = ", crits)

# Find increasing/decreasing intervals
if fp.subs(x, -4) > 0:
  print(f"f is increasing on (-inf, {crits[1]})")
else:
  print(f"f is decreasing on (-inf, {crits[1]})")
if fp.subs(x, -3) > 0:
  print(f"f is increasing on ({crits[1]}, {crits[2]})")
else:
  print(f"f is decreasing on ({crits[1]}, {crits[2]})")
if fp.subs(x, -2) > 0:
  print(f"f is increasing on ({crits[2]}, {crits[0]})")
else:
  print(f"f is decreasing on ({crits[2]}, {crits[0]})")
if fp.subs(x, 0.5) > 0:
  print(f"f is increasing on ({crits[0]}, {crits[3]})")
else:
  print(f"f is decreasing on ({crits[0]}, {crits[3]})")
if fp.subs(x, 0.66) > 0:
  print(f"f is increasing on ({crits[3]}, {crits[4]})")
else:
  print(f"f is decreasing on ({crits[3]}, {crits[4]})")
if fp.subs(x, 1) > 0:
  print(f"f is increasing on ({crits[4]}, {crits[5]})")
else:
  print(f"f is decreasing on ({crits[4]}, {crits[5]})")
if fp.subs(x, 3) > 0:
  print(f"f is increasing on ({crits[5]}, inf)")
else:
  print(f"f is decreasing on ({crits[5]}, inf)")

# Plot f'
sp.plot(fp, xlim=[-4, 4], ylim=[-4, 4])
```

#### 1b

```python
# Find points of inflection
fpp = sp.diff(fp, x)
inflects = []
for inflect in sp.solve(fpp, x):
  inflects.append(inflect.evalf())

print("Inflection points at x = ", inflects)

# Find concave up/down intervals
if fpp.subs(x, -3) > 0:
  print(f"f is concave up on (-inf, {inflects[0]})")
else:
  print(f"f is concave down on (-inf, {inflects[0]})")
if fpp.subs(x, -2) > 0:
  print(f"f is concave up on ({inflects[0]}, {inflects[1]})")
else:
  print(f"f is concave down on ({inflects[0]}, {inflects[1]})")
if fpp.subs(x, 0) > 0:
  print(f"f is concave up on ({inflects[1]}, {inflects[2]})")
else:
  print(f"f is concave down on ({inflects[1]}, {inflects[2]})")
if fpp.subs(x, 0.5) > 0:
  print(f"f is concave up on ({inflects[2]}, {inflects[3]})")
else:
  print(f"f is concave down on ({inflects[2]}, {inflects[3]})")
if fpp.subs(x, 1) > 0:
  print(f"f is concave up on ({inflects[3]}, {inflects[4]})")
else:
  print(f"f is concave down on ({inflects[3]}, {inflects[4]})")
if fpp.subs(x, 3) > 0:
  print(f"f is concave up on ({inflects[4]}, inf)")
else:
  print(f"f is concave down on ({inflects[4]}, inf)")

# Plot f''
sp.plot(fpp, xlim=[-4, 4], ylim=[-4, 4])
```

### Question 3

#### 3a

```python
sigma, x = sp.symbols('σ x')

f = ( 1 / (sigma * (2 * sp.pi) ** 0.5) ) * sp.E ** ( sp.Rational(-1, 2) * (x / sigma)**2 )

for sigma_v in [1, 5, 10]:
    F = f.subs(sigma, sigma_v)
    dFdx = sp.diff(F, x)
    cvs = sp.solve(dFdx)
    print(f'When σ = {sigma_v}, the critical value is {cvs[0]:.2f}')
```

#### 3b

```python
for sigma_v in [1, 5, 10]:
    F = f.subs(sigma, sigma_v)
    dFdx = sp.diff(F, x, 2)
    infpts_x = sp.solve(dFdx)
    infpts_y = [F.subs(x, x_v) for x_v in infpts_x]
    print(f'When σ = {sigma_v}, the inflection points are at ({infpts_x[0]:.2f}, {infpts_y[0]:.2f}) and ({infpts_x[1]:.2f}, {infpts_y[0]:.2f})')
```

The pattern is that, for a given σ, the x values that the inflection points occur at are σ and -σ.
